```python
import pandas as pd
import altair as alt
class visualizer():
    def __init__(self):
        pass
        
#     def search_player(self):
#         names = ['name1', 'name2']
#         rs = []
#         for i in range(len(names)): 
#             names[i] = input('Please enter player name: ').capitalize()

#             rs.append(df[ df['player.name'].str.contains(f'.*{names[i]}.*') ])
#             if len( rs[i] ) >= 1:
#                 print(rs[i][ ['player.name', 'player.id' ]])

#                 player_id = int(input('Specify player id: '))
#                 rs[i] = rs[i][ rs[i]['player.id'] == player_id ]
       
#         return pd.concat(rs)

    def search(self):
        name = input('name: ').capitalize()
        year = int(input("Season: "))

        rs = df[ df['player.name'].str.contains(f'.*{name}.*') and df["league.season"].equals(year) ]
#         if len( rs ) != 1:
#             print(rs[ ['player.name', 'player.id' ]])

#             player_id = int(input('player id: '))
#             rs = rs[ rs['player.id'] == player_id ]
#             return rs
#         else:
        return rs
    def search_players(self):
        results = ['','']
       
        for i in range(2):
            results[i] = self.search()
            while len(results[i]) <1:
                print("player not found, please try again")
                results[i] = self.search()
            if len(results[i]) >1:
                print(results[i][ ['player.name', 'player.id' ]])

                player_id = int(input('specify player id: '))
                results[i] = results[i][ results[i]['player.id'] == player_id ]
            print("registered: {}".format(results[i]["player.name"].tolist()))
        return pd.concat(results)
        
    
    def process_df(self,rs_df):
        positions = []
    
        for index, row in rs_df.iterrows():
            positions.append(row['games.position'])
        if positions[0] == positions[1]:
            if positions[0] == 'Attacker':
                rs_df = rs_df[['player.lastname','games.lineups', 'games.appearences', 'games.rating','goals.total', 'fouls.committed',
                        'goals.assists', 'shots.on']]
            elif positions[0] == 'Defender':
                rs_df = rs_df[['player.lastname','games.lineups', 'games.appearences',  'games.rating','goals.total', 'fouls.committed',
                    'goals.assists', 'tackles.total']]
            elif positions[0] == 'Midfielder':
                rs_df = rs_df[['player.lastname','games.lineups', 'games.appearences',  'games.rating','goals.total','fouls.committed',
                    'goals.assists', 'passes.key']]
            elif positions[0] == 'Goalkeeper':
                rs_df = rs_df[['player.lastname','games.lineups', 'games.appearences', 'games.rating','penalty.saved','fouls.committed',
                    'goals.conceded', 'goals.saves']]
        else: 
            rs_df = rs_df[['player.lastname','games.lineups', 'games.appearences', 'games.rating','goals.total','fouls.committed',
                       'goals.assists']]
            
        rs_df = rs_df.melt(id_vars=["player.lastname"], 
                var_name="stats", 
                value_name="value")

        
        return rs_df
    
    def visualize(self,vis_df):
    
        plot = alt.Chart(vis_df).mark_bar(opacity=0.4).encode(
            alt.X('value:Q', stack=False),
            alt.Y('stats:N'),
            alt.Color('player\.lastname:N', title='player'))
        return plot
```
